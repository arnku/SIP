```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
```

# Pixel-wise contrast enhancement

```python
IMG = skimage.io.imread('image.png')
IMG_GREY = skimage.io.imread('image.png', as_gray=True)

plt.figure(1)
plt.imshow(IMG)
plt.axis('off')
plt.figure(2)
plt.imshow(IMG_GREY, cmap='gray')
plt.axis('off')
plt.show()
```

## 1.1

```python
def gamma_transform(image, gamma):
    img_gamma = image ** gamma
    img_gamma = np.clip(img_gamma, 0, 1)
    return img_gamma
```

```python
img_f = skimage.util.img_as_float(IMG_GREY)

for i, gamma in enumerate([0.5, 1.0, 2.0]):
    plt.figure(i)
    plt.axis('off')
    img_gamma = gamma_transform(img_f, gamma)
    plt.imshow(img_gamma, cmap='gray')

plt.show()
```

## 1.2

```python
img = skimage.io.imread('autumn.tif')
img_f = skimage.util.img_as_float(img)

plt.imshow(img_f)
plt.axis('off')
plt.show()
```

```python
img_r = img_f[:, :, 0]
img_g = img_f[:, :, 1]
img_b = img_f[:, :, 2]

gamma = 0.5
img_r_gamma = gamma_transform(img_r, gamma)
img_g_gamma = gamma_transform(img_g, gamma)
img_b_gamma = gamma_transform(img_b, gamma)

img_gamma = np.stack((img_r_gamma, img_g_gamma, img_b_gamma), axis=2)
plt.imshow(img_gamma)
plt.axis('off')
plt.show()
```

# 1.3

```python
img_hsv = skimage.color.rgb2hsv(img_f)
img_h = img_hsv[:, :, 0]
img_s = img_hsv[:, :, 1]
img_v = img_hsv[:, :, 2]

gamma = 0.5
img_v_gamma = gamma_transform(img_v, gamma)

img_hsv_gamma = np.stack((img_h, img_s, img_v_gamma), axis=2)
img_rgb_gamma = skimage.color.hsv2rgb(img_hsv_gamma)
plt.imshow(img_rgb_gamma)
plt.axis('off')
plt.show()
```
